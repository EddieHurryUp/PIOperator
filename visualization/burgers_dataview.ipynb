```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import loadmat
import h5py
```

```python
plt.rcParams['font.size'] = 30
# view the ns equation and print some useful information
file_path = 'E:\\dfno\\PIOperator\\data\\burgers\\2d\\burgers_2d.mat'
file = loadmat(file_path)
print(list(file.keys()))
```

```python
key_turple = ('gamma', 'input', 'output', 'sigma', 'tau', 'tspan')
for key in key_turple:
    print(f'{key} shape:{np.array(file[key]).shape}')
```

code for viewing the relative field

```python
# begin record the value from the file
output = np.array(file['output'])  # shape:(1024, 241, 241) the vorticity at time:1 to time:50
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
P_test = 101
t = np.linspace(0, 1, P_test)
x = np.linspace(0, 1, P_test)
T, X = np.meshgrid(t, x)
plt.pcolor(T, X, output[500, :, :,], cmap='jet')
plt.xlabel('$x$')
plt.ylabel('$t$')
plt.title('Exact $u(x,t)$')
plt.colorbar()
plt.tight_layout()
plt.savefig('burges_1d', dpi=600)
```
