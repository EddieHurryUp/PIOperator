```python
import pandas as pd  
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt  

# x, y, metrics(x,y) axis setting
learning_rates = [round(rate, 5) for rate in np.linspace(5e-4, 4e-3, 10)]
model_sizes = np.array(range(11, 21))
test_loss = np.load('E:\\dfno\\PIOperator\\visualization\\test_error.npy')
  
# # generate pd data 
data = pd.DataFrame(test_loss, columns=model_sizes, index=learning_rates)  

# filter some value
# data[data > 2.] = 2.
  
# drop the duplicated data  
# data = data.drop_duplicates(subset=['Learning Rate', 'Model Size'])  
 
# # transform 
# pivot_data = data.pivot("Learning Rate", "Model Size", "Test Loss")  
  
# plot the heatmap  
plt.figure(figsize=(12, 10))  
# cmap="YlGnBu"
cmap='coolwarm'
# cmap = sns.color_palette("cubehelix", as_cmap=True) 
heatmap = sns.heatmap(data, annot=True, fmt=".5f", cmap=cmap, linewidths=0.5, linecolor='grey')   
  
# set title and axis 
plt.title('Test Loss for Different Learning Rates and Model Sizes', fontsize=16)  
plt.xlabel('Model Size', fontsize=12)  
plt.ylabel('Learning Rate', fontsize=12)  
  
# set font  
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10, rotation=0)  

# set tick
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True, length=6, pad=7)  
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=True, length=6, pad=7) 

# save figure and show in this window
plt.tight_layout()
plt.savefig('heatmap_test.jpg', dpi=600)
plt.show()  
```
