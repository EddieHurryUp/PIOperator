```python
import pandas as pd  
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt  

# generate rand data 
num_points = 200  
learning_rates = np.round(np.random.uniform(0.001, 0.01, num_points), 3)  
model_sizes = np.random.choice(np.array(range(11, 21)), num_points)  
test_loss = np.round(np.random.uniform(0.1, 1.0, num_points), 3)  
  
# generate pd data 
data = pd.DataFrame({'Learning Rate': learning_rates, 'Model Size': model_sizes, 'Test Loss': test_loss})  
  
# drop the duplicated data  
data = data.drop_duplicates(subset=['Learning Rate', 'Model Size'])  
  
# transform 
pivot_data = data.pivot("Learning Rate", "Model Size", "Test Loss")  
  
# plot the heatmap  
plt.figure(figsize=(12, 10))  
# cmap="YlGnBu"
cmap='coolwarm'
# cmap = sns.color_palette("cubehelix", as_cmap=True) 
heatmap = sns.heatmap(pivot_data, annot=True, fmt=".3f", cmap=cmap, linewidths=0.5, linecolor='grey')   
  
# set title and axis 
plt.title('Test Loss for Different Learning Rates and Model Sizes', fontsize=16)  
plt.xlabel('Model Size', fontsize=12)  
plt.ylabel('Learning Rate', fontsize=12)  
  
# set font  
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10)  

# set tick
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True, length=6, pad=7)  
plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=True, length=6, pad=7) 

# save figure and show in this window
plt.tight_layout()
plt.savefig('heatmap_test.jpg', dpi=600)
plt.show()  
```
